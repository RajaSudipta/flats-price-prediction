Comments: Price are in different units. 'Lac' and 'Crore'. So, bringing all of the datapoints into same unit 'Crore'

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```
